```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv("C:\\Users\\hp\\Desktop\\apriopri\\Market_Basket_Optimisation.csv", header=None)
```

```python
df.head()
#df.shape (7501,20)
```

```python
#df.fillna(0, inplace=True)
'''aprori needs data in list format for transactions'''
transactions= []
for i in range(0, len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j]) !="0"] )        
```

```python
transactions[1]
```

```python
df.values[1,2]
```

```python
#apriori called wrt support , confidence and lift
## min_support = 0.003 -> means selecting items with min support of 0.3%
## min_confidance = 0.2 -> means min confidance of 20% 
## min_lift = 3  
## min_length = 2 -> means no. of items in the transaction should be 2 


rules= apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift= 3,
               min_length= 2)
rules
```

```python
df_results= pd.DataFrame(rules)
df_results
```

```python
support= df_results.support
```

```python
support
```

```python
df_results["ordered_statistics"][0]
```

```python
first_value=  []
second_value= []
third_value=  []
fourth_value= []
for i in range(df_results.shape[0]):
    single_list= df_results["ordered_statistics"][i][0]
    first_value.append(list(single_list[0]))
    second_value.append(list(single_list[1]))
    third_value.append(single_list[2])
    fourth_value.append(single_list[3])
```

```python
lhs= pd.DataFrame(first_value)
rhs= pd.DataFrame(second_value)
confidence= pd.DataFrame(third_value, columns=["Confidence"])
lift= pd.DataFrame(fourth_value, columns=["Lift"])
df_final= pd.concat([lhs, rhs, support, confidence, lift], axis= 1)
```
