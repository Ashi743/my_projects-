#loading  the datsets

```python
gen1= pd.read_csv("/root/.cache/kagglehub/datasets/anikannal/solar-power-generation-data/versions/1/Plant_1_Generation_Data.csv")
gen2= pd.read_csv("/root/.cache/kagglehub/datasets/anikannal/solar-power-generation-data/versions/1/Plant_2_Generation_Data.csv")
weather1= pd.read_csv("/root/.cache/kagglehub/datasets/anikannal/solar-power-generation-data/versions/1/Plant_1_Weather_Sensor_Data.csv")
weather2= pd.read_csv("/root/.cache/kagglehub/datasets/anikannal/solar-power-generation-data/versions/1/Plant_2_Weather_Sensor_Data.csv")
```

```python
gen1.head()
```

change date to datetime format

```python

gen1['DATE_TIME'] = pd.to_datetime(gen1['DATE_TIME'])
gen2['DATE_TIME'] = pd.to_datetime(gen2['DATE_TIME'])
weather1['DATE_TIME'] = pd.to_datetime(weather1['DATE_TIME'])
weather2['DATE_TIME'] = pd.to_datetime(weather2['DATE_TIME'])


gen1['DATE_TIME'] = gen1['DATE_TIME'].dt.strftime('%Y-%m-%d %H:%M')
gen2['DATE_TIME'] = gen2['DATE_TIME'].dt.strftime('%Y-%m-%d %H:%M')
weather1['DATE_TIME'] = weather1['DATE_TIME'].dt.strftime('%Y-%m-%d %H:%M')
weather2['DATE_TIME'] = weather2['DATE_TIME'].dt.strftime('%Y-%m-%d %H:%M')
```

```python
df_solar1 = pd.merge(gen1.drop(columns = ['PLANT_ID']), weather1.drop(columns = ['PLANT_ID', 'SOURCE_KEY']), on='DATE_TIME')
df_solar1.sample(5).style.background_gradient(cmap='cool')
```

```python
df_solar2 = pd.merge(gen2.drop(columns = ['PLANT_ID']), weather2.drop(columns = ['PLANT_ID', 'SOURCE_KEY']), on='DATE_TIME')
df_solar2.sample(5).style.background_gradient(cmap='cool')
```

```python
# adding separate time and date columns
df_solar1["DATE"] = pd.to_datetime(df_solar1["DATE_TIME"]).dt.date
df_solar1["TIME"] = pd.to_datetime(df_solar1["DATE_TIME"]).dt.time
df_solar1['DAY'] = pd.to_datetime(df_solar1['DATE_TIME']).dt.day
df_solar1['MONTH'] = pd.to_datetime(df_solar1['DATE_TIME']).dt.month
df_solar1['WEEK'] = pd.to_datetime(df_solar1['DATE_TIME']).dt.isocalendar().week # Use .dt.isocalendar().week to get the week number


# add hours and minutes for ml models
df_solar1['HOURS'] = pd.to_datetime(df_solar1['TIME'],format='%H:%M:%S').dt.hour
df_solar1['MINUTES'] = pd.to_datetime(df_solar1['TIME'],format='%H:%M:%S').dt.minute
df_solar1['TOTAL MINUTES PASS'] = df_solar1['MINUTES'] + df_solar1['HOURS']*60

# add date as string column
df_solar1["DATE_STRING"] = df_solar1["DATE"].astype(str) # add column with date as string
df_solar1["HOURS"] = df_solar1["HOURS"].astype(str)
df_solar1["TIME"] = df_solar1["TIME"].astype(str)
```

```python
# Adding separate time and date columns
df_solar2["DATE"] = pd.to_datetime(df_solar2["DATE_TIME"]).dt.date
df_solar2["TIME"] = pd.to_datetime(df_solar2["DATE_TIME"]).dt.time
df_solar2['DAY'] = pd.to_datetime(df_solar2['DATE_TIME']).dt.day
df_solar2['MONTH'] = pd.to_datetime(df_solar2['DATE_TIME']).dt.month
df_solar2['WEEK'] = pd.to_datetime(df_solar2['DATE_TIME']).dt.isocalendar().week  # Use .dt.isocalendar().week to get the week number

# Add hours and minutes for ML models
df_solar2['HOURS'] = pd.to_datetime(df_solar2['TIME'], format='%H:%M:%S').dt.hour
df_solar2['MINUTES'] = pd.to_datetime(df_solar2['TIME'], format='%H:%M:%S').dt.minute
df_solar2['TOTAL MINUTES PASS'] = df_solar2['MINUTES'] + df_solar2['HOURS'] * 60

# Add date as string column
df_solar2["DATE_STRING"] = df_solar2["DATE"].astype(str)  # Add column with date as string
df_solar2["HOURS"] = df_solar2["HOURS"].astype(str)
df_solar2["TIME"] = df_solar2["TIME"].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder2 = LabelEncoder()
df_solar2['SOURCE_KEY_NUMBER'] = encoder2.fit_transform(df_solar2['SOURCE_KEY'])
df_solar2.head()
```
